```python
import json
import requests
import pandas as pd
```

```python
# Charger les données JSON à partir du fichier local
with open('../datas/panorama-des-festivals.json', 'r', encoding='utf-8') as file:
    data_json = json.load(file)
```

```python
# Charger les données JSON dans un DataFrame pandas
paca_festivals_raw = pd.DataFrame(data_json)
```

```python
# Extraction des données de la colonne "fields" et convertit en database
paca_festivals = pd.json_normalize(paca_festivals_raw["fields"])
display(paca_festivals.tail())
```

#### Quel est le mois de l'année le plus chargé en festivals ?

```python
from calendar import month_name

def month_from_number(number):
    if 1 <= number <= 12:
        return month_name[number]

best_month_float = paca_festivals.groupby('mois_indicatif_en_chiffre_y_compris_double_mois')['mois_indicatif_en_chiffre_y_compris_double_mois'].count().idxmax()
best_month_number = int(best_month_float)

best_month = month_from_number(best_month_number)

#result is in english but we are working in french, to translate

from googletrans import Translator

def traduire_texte(texte, langue_destination='fr'):
    translator = Translator()
    traduction = translator.translate(texte, dest=langue_destination)
    return traduction.text

print('C\'est le mois de', traduire_texte(best_month),'.')
```
